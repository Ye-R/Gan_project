```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# KNN, SVM용 Labeling : 정상, inner, outer 순

```python
NoOfRealNormal = 628
NoOfRealMisalign = 382
NoOfFakeNormal = 300
NoOfFakeMisalign = 300
```

```python
Training_All_Label = np.zeros((NoOfFakeNormal+NoOfFakeMisalign,1))
Training_All_Label[:NoOfFakeNormal]=1
Training_All_Label[NoOfFakeNormal:]=0


Training_All_Label = pd.DataFrame(Training_All_Label)
```

```python
Training_All_Label
```

```python
Test_Label = np.zeros((NoOfRealNormal+NoOfRealMisalign,1))
Test_Label[:NoOfRealNormal]=1
Test_Label[NoOfRealNormal:NoOfRealNormal+NoOfRealMisalign]=0


Test_Label = pd.DataFrame(Test_Label)
```

```python
Test_Label
```

## ANN용 One-Hot Encoding : 정상, inner, outer 순

```python
Training_All_Label_forANN = np.zeros((NoOfFakeNormal+NoOfFakeMisalign,2))

Training_All_Label_forANN[:NoOfFakeNormal , 0] = 1
Training_All_Label_forANN[NoOfFakeNormal:NoOfFakeNormal+NoOfFakeMisalign , 1] = 1

Training_All_Label_forANN = pd.DataFrame(Training_All_Label_forANN)

pd.set_option('display.max_row', 600)
pd.set_option('display.max_columns', 100)
Training_All_Label_forANN
```

```python
Test_Label_forANN = np.zeros((NoOfRealNormal+NoOfRealMisalign,2))

Test_Label_forANN[:NoOfRealNormal , 0] = 1
Test_Label_forANN[NoOfRealNormal:NoOfRealNormal+NoOfRealMisalign , 1] = 1

Test_Label_forANN = pd.DataFrame(Test_Label_forANN)

pd.set_option('display.max_row', 1268)
pd.set_option('display.max_columns', 100)

Test_Label_forANN
```
